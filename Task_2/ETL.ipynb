```python
if "Unnamed: 0" in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python
df['Installs'] = df['Installs'].astype(str)                            
df['Installs'] = df['Installs'].str.replace(r'[^\d]', '', regex=True) 
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')       
df['Installs'] = df['Installs'].fillna(0).astype(int)                 

df['Price'] = df['Price'].astype(str).replace('Free', '0')
df['Price'] = df['Price'].str.replace(r'[^\d\.]', '', regex=True)  # Remove ₹, $, etc.
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price'] = df['Price'].fillna(0).astype(int)
```

```python
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
```

```python
#
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
df['Size'] = df['Size'].fillna(df['Size'].median())

df['Current Ver'] = df['Current Ver'].fillna('Unknown')
df['Android Ver'] = df['Android Ver'].fillna('Unknown')
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
#Creating Random countries because 
# we dont have countries column for preparing choropleth
countries=[
    'India','USA','Brazil','Germany','Japan'
]
df['Country']=np.random.choice(countries, size=len(df))
```

```python
# Step 1: Exclude categories starting with A, C, G, S
df_filtered = df[~df['Category'].str.startswith(('A', 'C', 'G', 'S'))]

# Step 2: Group by category and sum installs
category_installs = df_filtered.groupby('Category')['Installs'].sum()

# Step 3: Get top 5 categories by installs
top5_categories = category_installs.sort_values(ascending=False).head(5)

# Step 4: Mark rows with installs > 1,000,000
df['HighInstall'] = df['Installs'] > 1000000
```
